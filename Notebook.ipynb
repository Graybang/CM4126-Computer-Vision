```python
# import necessary modules 
import cv2 as cv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
from PIL import Image
import requests
from io import BytesIO
```

```python
# open and display image
url = 'https://github.com/Graybang/CM4126-Computer-Vision/blob/main/image.jfif?raw=true'

response = requests.get(url)

image = Image.open(BytesIO(response.content))


plt.imshow(image)
```

```python
# convert the image into an array
image_arr = np.array(image)
# check the array shape
image_arr.shape
```

```python
gray = cv.cvtColor(image_arr,cv.COLOR_BGR2GRAY)
#show the gray image
plt.imshow(gray, cmap='gray')
```

```python
red_channel = image_arr[:,:,0]
green_channel = image_arr[:,:,1]
blue_channel = image_arr[:,:,2]
```

```python
plt.figure(figsize=[18,5])
plt.subplot(131);plt.imshow(red_channel);plt.title('Red Channel')
plt.subplot(132);plt.imshow(green_channel);plt.title('Green Channel')
plt.subplot(133);plt.imshow(blue_channel);plt.title('Blue Channel')
```

```python
#open video file
vcap = cv.VideoCapture('https://github.com/Graybang/CM4126-Computer-Vision/blob/main/LoathsomeMellowAmphibian-mobile.mp4?raw=true')

# how many frames in the file
length = int(vcap.get(cv.CAP_PROP_FRAME_COUNT))
i=0
# sample every 50 frame and save as imaes
while(vcap.isOpened()):
    ret, frame = vcap.read()
    if ret == False:
        break
    if i%50==0:
        cv.imwrite('Data/'+str(i)+'.jpg',frame)
    i+=1

vcap.release()
cv.destroyAllWindows()
```

```python
import glob
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
%matplotlib inline

images = []
for img_path in glob.glob('Data/*.jpg'):
    images.append(mpimg.imread(img_path))

plt.figure(figsize=(20,10))
columns = 5
for i, image in enumerate(images):
    plt.subplot(int(len(images) / columns + 1), columns, i + 1)
    plt.imshow(image)
```

```python
def extractImages(pathIn, pathOut,number_of_frames):
  count = 0
  down_width = 100
  down_height = 100
  down_points = (down_width, down_height)
  vidcap = cv.VideoCapture(pathIn)
  success,image = vidcap.read()
  success = True
  while success and count<number_of_frames:
    vidcap.set(cv.CAP_PROP_POS_MSEC,(count * 4))    # changed this line to match the ms per frame of the video
    success,image = vidcap.read()
    image = cv.resize(image, down_points, interpolation= cv.INTER_LINEAR)
    print ('Read a new frame and save it: ', count, ' ',success)
    cv.imwrite( pathOut + "frame%d.jpg" % count, image)     # save frame as JPEG file
    count = count + 1
```

```python
video_path = 'https://github.com/Graybang/CM4126-Computer-Vision/blob/main/LoathsomeMellowAmphibian-mobile.mp4?raw=true'
images_path = 'Data/VideoImages/'
```

```python
extractImages(video_path, images_path,25)
```
